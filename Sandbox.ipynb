```python
import pandas as pd
import torch
from transformers import BertTokenizerFast

# Load the CSV file
df = pd.read_csv('/content/lesson_script_variations(1).csv')

# Ensure all inputs are strings
df['context'] = df['context'].astype(str)
df['question'] = df['question'].astype(str)
df['answer'] = df['answer'].astype(str)

# Initialize the tokenizer
tokenizer = BertTokenizerFast.from_pretrained('bert-base-uncased')

# Tokenize the data
def tokenize_function(row):
    return tokenizer(row['context'], row['question'], truncation=True, padding='max_length', max_length=512)

# Apply the function to each row in the DataFrame and collect the results in lists
input_ids = []
attention_mask = []
for idx, row in df.iterrows():
    tokens = tokenize_function(row)
    input_ids.append(tokens['input_ids'])
    attention_mask.append(tokens['attention_mask'])
```

```python
def find_answer_positions(context, answer, tokenizer):
    tokenized_context = tokenizer(context, return_offsets_mapping=True)
    context_tokens = tokenized_context['input_ids']
    offset_mapping = tokenized_context['offset_mapping']

    tokenized_answer = tokenizer(answer, add_special_tokens=False)
    answer_tokens = tokenized_answer['input_ids']

    for idx in range(len(context_tokens) - len(answer_tokens) + 1):
        if context_tokens[idx:idx + len(answer_tokens)] == answer_tokens:
            start_position = idx
            end_position = idx + len(answer_tokens) - 1
            return start_position, end_position

    return None, None

# Apply the function to each row in the DataFrame
df['start_position'], df['end_position'] = zip(*df.apply(lambda row: find_answer_positions(row['context'], row['answer'], tokenizer), axis=1))
```

```python
# Filter out rows where start or end positions could not be found
df = df.dropna(subset=['start_position', 'end_position'])

# Ensure start and end positions are integers
df['start_position'] = df['start_position'].astype(int)
df['end_position'] = df['end_position'].astype(int)

# Convert tokenized data to tensors
input_ids = torch.tensor(input_ids)
attention_mask = torch.tensor(attention_mask)
start_positions = torch.tensor(df['start_position'].tolist())
end_positions = torch.tensor(df['end_position'].tolist())
labels = torch.tensor(df['answer'].tolist())
```

```python
class CustomDataset(torch.utils.data.Dataset):
    def __init__(self, input_ids, attention_mask, labels, start_positions, end_positions):
        self.input_ids = input_ids
        self.attention_mask = attention_mask
        self.labels = labels
        self.start_positions = start_positions
        self.end_positions = end_positions

    def __getitem__(self, idx):
        return {
            'input_ids': self.input_ids[idx],
            'attention_mask': self.attention_mask[idx],
            'labels': self.labels[idx],
            'start_positions': self.start_positions[idx],
            'end_positions': self.end_positions[idx],
        }

    def __len__(self):
        return len(self.labels)

# Create the dataset
dataset = CustomDataset(input_ids, attention_mask, labels, start_positions, end_positions)
```
